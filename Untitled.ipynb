```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('dynamic_photofeeler.db')

# Exécuter une requête SQL
query = "SELECT * FROM Photos"
df_photos = pd.read_sql_query(query, conn)
print("Photos:")
print(df_photos)

query = "SELECT * FROM Scores"
df_scores = pd.read_sql_query(query, conn)
print("\nScores:")
print(df_scores)

query = "SELECT * FROM Votes"
df_votes = pd.read_sql_query(query, conn)
print("\nVotes:")
print(df_votes)

# Fermer la connexion
conn.close()
```
